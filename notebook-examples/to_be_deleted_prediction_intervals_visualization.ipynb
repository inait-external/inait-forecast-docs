# ğŸ¯ Creating Ensemble Predictions with Rich Uncertainty Quantification

In this section, we'll create ensemble predictions specifically designed to showcase **prediction intervals** and **uncertainty quantification**. We'll use our champion ensemble combination to ensure the highest quality confidence bounds.


## ğŸ“Š Enhanced Configuration for Uncertainty Analysis

We'll optimize our parameters for the best uncertainty quantification:

- **Models**: `"neural,gradient_boost"` - Champion uncertainty combination
- **Prediction Intervals**: Automatically generated with ensemble
- **Confidence Levels**: Multiple levels (80%, 90%, 95%)
- **Uncertainty Sources**: Both epistemic and aleatoric uncertainty
- **Visualization Ready**: Structured for advanced plotting


```python
# ğŸ¯ Configure ensemble prediction for optimal uncertainty quantification
prediction_url = base_url + "/prediction"
data_path = "../data/demo_power.csv"
target_columns = "DE_Spot_EPEX_1H_A"  # German electricity spot prices
features = "DE_Residual_Load_15M_A_AVG,DE_Consumption_15M_A_AVG"
forecasting_horizon = 15  # Longer horizon for better uncertainty visualization
observation_length = 24  # Rich historical context for uncertainty estimation

# ğŸ† Champion ensemble for uncertainty quantification
ensemble_models = "neural,gradient_boost"

print("ğŸ¯ Prediction Intervals Configuration:")
print(f"ğŸ“Š Dataset: {data_path}")
print(f"ğŸ¯ Target: {target_columns} (electricity spot prices)")
print(f"ğŸ”® Forecast Horizon: {forecasting_horizon} time steps")
print(f"ğŸ“ˆ Historical Context: {observation_length} data points")
print(f"ğŸ¤– Ensemble Models: {ensemble_models}")
print("ğŸ¨ Focus: Rich prediction intervals and uncertainty bands")

print("\nğŸ”¬ Uncertainty Quantification Strategy:")
print("   ğŸ§  Neural Network: Epistemic uncertainty from deep learning")
print("   ğŸŒ³ Gradient Boost: Aleatoric uncertainty from ensemble trees")
print("   ğŸ¤ Combination: Comprehensive uncertainty estimation")
print("   ğŸ“Š Output: Multi-level confidence intervals")

# ğŸ“¦ Create payload optimized for prediction intervals
payload = create_payload_from_file(
    file_path=data_path,
    target_columns=target_columns,
    feature_columns=features,
    forecasting_horizon=forecasting_horizon,
    observation_length=observation_length,
    models=ensemble_models,
    prediction_interval_levels="50",
    background=True,  # Essential for quality uncertainty quantification
)

print("\nğŸ“¤ Submitting ensemble request for uncertainty analysis...")
print("âš¡ Processing ensemble with prediction intervals...")

# ğŸš€ Submit ensemble prediction request
response = make_request(prediction_url, payload, auth_key=auth_key)

print("âœ… Ensemble submission successful!")
print(f"ğŸ†” Session ID: {response['session_id']}")
print("ğŸ”„ Status: Background processing initiated")

# Store session_id for monitoring
session_id = response["session_id"]
status_url = base_url + "/status/"

# ğŸ“Š Monitor ensemble processing with appropriate intervals
print("\nğŸ”„ Monitoring ensemble processing for uncertainty quantification...")

# Initial status check
status_response = make_get_request(status_url, session_id=session_id, auth_key=auth_key)
print(f"ğŸ“Š Initial Status: {status_response['status']}")

# Optimized polling for uncertainty-focused ensemble
poll_count = 0
start_time = time.time()

while status_response["status"] == "in_progress":
    poll_count += 1
    elapsed = time.time() - start_time
    wait_time = min(
        6 + (poll_count * 2), 18
    )  # Appropriate for uncertainty calculations

    print(
        f"â³ Poll #{poll_count}: Processing uncertainty quantification... (Elapsed: {elapsed:.1f}s)"
    )
    print("   ğŸ§  Neural: Computing epistemic uncertainty bounds")
    print("   ğŸŒ³ Gradient Boost: Calculating ensemble prediction intervals")
    print(f"   â±ï¸  Next check in {wait_time} seconds...")

    time.sleep(wait_time)
    status_response = make_get_request(
        status_url, session_id=session_id, auth_key=auth_key
    )

# ğŸ‰ Process completion
total_time = time.time() - start_time

if status_response["status"] == "completed":
    print("\nğŸŠ Ensemble with uncertainty quantification completed!")
    print(f"â±ï¸  Total processing time: {total_time:.1f} seconds")
    print(f"ğŸ“Š Polling cycles: {poll_count}")

    # ğŸ“¥ Retrieve results with prediction intervals
    result_url = base_url + "/result/"
    result_response = make_get_request(
        result_url, session_id=session_id, auth_key=auth_key
    )

    print("ğŸ“Š Uncertainty data retrieved successfully!")

    # Extract DataFrame with prediction intervals
    df_uncertainty, _ = get_dataframe_from_response(result_response["result"])

    print("\nğŸ¯ Prediction Intervals Dataset:")
    print(f"   ğŸ“ˆ Shape: {df_uncertainty.shape}")
    print(f"   ğŸ¯ Columns: {list(df_uncertainty.columns)}")

    success = True

elif status_response["status"] == "failed":
    print(f"âŒ Ensemble failed: {status_response.get('error', 'Unknown error')}")
    df_uncertainty = None
    success = False
else:
    print(f"âš ï¸  Unexpected status: {status_response['status']}")
    df_uncertainty = None
    success = False

df_uncertainty
```

## ğŸ“Š **Built-in API Plot**
We'll use the Inait Predict API's built-in plotting to see their default uncertainty visualization.


```python
# ğŸ–¼ï¸ Generate built-in API plot with prediction intervals
if success and session_id:
    print("ğŸ¨ Generating built-in API visualization with uncertainty...")

    plot_url = base_url + "/plot"
    plot_type = "prediction"  # Request prediction plot with intervals

    print("ğŸ“Š API Plot Configuration:")
    print(f"   ğŸ†” Session: {session_id}")
    print(f"   ğŸ¯ Type: {plot_type}")
    print("   ğŸ“ˆ Focus: Prediction intervals and uncertainty bands")

    # Create plot payload
    plot_payload = create_plot_payload(
        session_id, plot_type, cutoff_date=None, prediction_for=None
    )

    print("\nğŸ“¤ Requesting API plot with uncertainty visualization...")

    try:
        plot_response = make_request(plot_url, plot_payload, auth_key=auth_key)

        if "response" in plot_response and "data" in plot_response["response"]:
            # Extract and display the plot
            image_bytes = bytes_from_base64(plot_response["response"]["data"])
            plot_image(image_bytes)

            print("\nâœ… Built-in API Plot Generated!")
            print("ğŸ¨ Features visible in the plot:")
            print("   ğŸ“ˆ Historical data (blue line)")
            print("   ğŸ”® Ensemble predictions (orange line)")
            print("   ğŸ“Š Prediction intervals (gray shaded bands)")
            print("   ğŸ¯ Automatic uncertainty quantification")

        else:
            print("âš ï¸  Built-in plot generation failed")
            print(f"Response: {plot_response}")

    except Exception as e:
        print(f"âš ï¸  Error generating built-in plot: {e}")
        print("ğŸ”„ Continuing with custom visualization...")

else:
    print("âš ï¸  Cannot generate built-in plot - no valid session")

print("\nğŸ¨ Proceeding to custom advanced visualization...")
```
