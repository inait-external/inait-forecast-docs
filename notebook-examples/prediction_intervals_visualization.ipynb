# üéØ Creating Ensemble Predictions with Rich Uncertainty Quantification

In this section, we'll create ensemble predictions specifically designed to showcase **prediction intervals** and **uncertainty quantification**. We'll use our champion ensemble combination to ensure the highest quality confidence bounds.


## üìä Enhanced Configuration for Uncertainty Analysis

We'll optimize our parameters for the best uncertainty quantification:

- **Models**: `"neural,gradient_boost"` - Champion uncertainty combination
- **Prediction Intervals**: Automatically generated with ensemble
- **Confidence Levels**: Multiple levels (80%, 90%, 95%)
- **Uncertainty Sources**: Both epistemic and aleatoric uncertainty
- **Visualization Ready**: Structured for advanced plotting


```python
# üéØ Configure ensemble prediction for optimal uncertainty quantification
prediction_url = base_url + "/prediction"
data_path = "../data/demo_power.csv"
target_columns = "DE_Spot_EPEX_1H_A"  # German electricity spot prices
features = "DE_Residual_Load_15M_A_AVG,DE_Consumption_15M_A_AVG"
forecasting_horizon = 15  # Longer horizon for better uncertainty visualization
observation_length = 24  # Rich historical context for uncertainty estimation

# üèÜ Champion ensemble for uncertainty quantification
ensemble_models = "neural,gradient_boost"

print("üéØ Prediction Intervals Configuration:")
print(f"üìä Dataset: {data_path}")
print(f"üéØ Target: {target_columns} (electricity spot prices)")
print(f"üîÆ Forecast Horizon: {forecasting_horizon} time steps")
print(f"üìà Historical Context: {observation_length} data points")
print(f"ü§ñ Ensemble Models: {ensemble_models}")
print("üé® Focus: Rich prediction intervals and uncertainty bands")

print("\nüî¨ Uncertainty Quantification Strategy:")
print("   üß† Neural Network: Epistemic uncertainty from deep learning")
print("   üå≥ Gradient Boost: Aleatoric uncertainty from ensemble trees")
print("   ü§ù Combination: Comprehensive uncertainty estimation")
print("   üìä Output: Multi-level confidence intervals")

# üì¶ Create payload optimized for prediction intervals
payload = create_payload_from_file(
    file_path=data_path,
    target_columns=target_columns,
    feature_columns=features,
    forecasting_horizon=forecasting_horizon,
    observation_length=observation_length,
    models=ensemble_models,
    prediction_interval_levels="50",
    background=True,  # Essential for quality uncertainty quantification
)

print("\nüì§ Submitting ensemble request for uncertainty analysis...")
print("‚ö° Processing ensemble with prediction intervals...")

# üöÄ Submit ensemble prediction request
response = make_request(prediction_url, payload, auth_key=auth_key)

print("‚úÖ Ensemble submission successful!")
print(f"üÜî Session ID: {response['session_id']}")
print("üîÑ Status: Background processing initiated")

# Store session_id for monitoring
session_id = response["session_id"]
status_url = base_url + "/status/"

# üìä Monitor ensemble processing with appropriate intervals
print("\nüîÑ Monitoring ensemble processing for uncertainty quantification...")

# Initial status check
status_response = make_get_request(status_url, session_id=session_id, auth_key=auth_key)
print(f"üìä Initial Status: {status_response['status']}")

# Optimized polling for uncertainty-focused ensemble
poll_count = 0
start_time = time.time()

while status_response["status"] == "in_progress":
    poll_count += 1
    elapsed = time.time() - start_time
    wait_time = min(
        6 + (poll_count * 2), 18
    )  # Appropriate for uncertainty calculations

    print(
        f"‚è≥ Poll #{poll_count}: Processing uncertainty quantification... (Elapsed: {elapsed:.1f}s)"
    )
    print("   üß† Neural: Computing epistemic uncertainty bounds")
    print("   üå≥ Gradient Boost: Calculating ensemble prediction intervals")
    print(f"   ‚è±Ô∏è  Next check in {wait_time} seconds...")

    time.sleep(wait_time)
    status_response = make_get_request(
        status_url, session_id=session_id, auth_key=auth_key
    )

# üéâ Process completion
total_time = time.time() - start_time

if status_response["status"] == "completed":
    print("\nüéä Ensemble with uncertainty quantification completed!")
    print(f"‚è±Ô∏è  Total processing time: {total_time:.1f} seconds")
    print(f"üìä Polling cycles: {poll_count}")

    # üì• Retrieve results with prediction intervals
    result_url = base_url + "/result/"
    result_response = make_get_request(
        result_url, session_id=session_id, auth_key=auth_key
    )

    print("üìä Uncertainty data retrieved successfully!")

    # Extract DataFrame with prediction intervals
    df_uncertainty, _ = get_dataframe_from_response(result_response["result"])

    print("\nüéØ Prediction Intervals Dataset:")
    print(f"   üìà Shape: {df_uncertainty.shape}")
    print(f"   üéØ Columns: {list(df_uncertainty.columns)}")

    success = True

elif status_response["status"] == "failed":
    print(f"‚ùå Ensemble failed: {status_response.get('error', 'Unknown error')}")
    df_uncertainty = None
    success = False
else:
    print(f"‚ö†Ô∏è  Unexpected status: {status_response['status']}")
    df_uncertainty = None
    success = False

df_uncertainty
```

#  üñºÔ∏è Visualization Strategy

## üìä **Built-in API Plot**
We'll use the Inait Predict API's built-in plotting to see their default uncertainty visualization.


```python
# üñºÔ∏è Generate built-in API plot with prediction intervals
if success and session_id:
    print("üé® Generating built-in API visualization with uncertainty...")

    plot_url = base_url + "/plot"
    plot_type = "prediction"  # Request prediction plot with intervals

    print("üìä API Plot Configuration:")
    print(f"   üÜî Session: {session_id}")
    print(f"   üéØ Type: {plot_type}")
    print("   üìà Focus: Prediction intervals and uncertainty bands")

    # Create plot payload
    plot_payload = create_plot_payload(
        session_id, plot_type, cutoff_date=None, prediction_for=None
    )

    print("\nüì§ Requesting API plot with uncertainty visualization...")

    try:
        plot_response = make_request(plot_url, plot_payload, auth_key=auth_key)

        if "response" in plot_response and "data" in plot_response["response"]:
            # Extract and display the plot
            image_bytes = bytes_from_base64(plot_response["response"]["data"])
            plot_image(image_bytes)

            print("\n‚úÖ Built-in API Plot Generated!")
            print("üé® Features visible in the plot:")
            print("   üìà Historical data (blue line)")
            print("   üîÆ Ensemble predictions (orange line)")
            print("   üìä Prediction intervals (gray shaded bands)")
            print("   üéØ Automatic uncertainty quantification")

        else:
            print("‚ö†Ô∏è  Built-in plot generation failed")
            print(f"Response: {plot_response}")

    except Exception as e:
        print(f"‚ö†Ô∏è  Error generating built-in plot: {e}")
        print("üîÑ Continuing with custom visualization...")

else:
    print("‚ö†Ô∏è  Cannot generate built-in plot - no valid session")

print("\nüé® Proceeding to custom advanced visualization...")
```
